#### Teste 1

Hipotese nula: A taxa de vitória de cada abertura não está relacionada ao rank dos jogadores.

Hipotese alternativa: A taxa de vitória de cada abertura está relacionada ao rank dos jogadores.

```python
# Players rating between 750 - 2500
# divide into groups by rating

# 750 - 1000
group_one = chess_df[(chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 1000]

# 1000 - 1250
group_two = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 1250)]
group_two = group_two[((group_two["white_rating"] + group_two["black_rating"]) / 2 > 1000)]

# 1250 - 1500
group_three = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 1500)]
group_three = group_three[((group_three["white_rating"] + group_three["black_rating"]) / 2 > 1250)]

# 1500 - 1750
group_four = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 1750)]
group_four = group_four[((group_four["white_rating"] + group_four["black_rating"]) / 2 > 1500)]

# 1750 - 2000
group_five = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 2000)]
group_five = group_five[((group_five["white_rating"] + group_five["black_rating"]) / 2 > 1750)]

# 2000 - 2250
group_six = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 <= 2250)]
group_six = group_six[((group_six["white_rating"] + group_six["black_rating"]) / 2 > 2000)]

# 2250 - 2500
group_seven = chess_df[((chess_df["white_rating"] + chess_df["black_rating"]) / 2 > 2250)]
```

```python
# openings with higher win rate by side

chess_winner_opening_df = chess_df.groupby(['winner','opening_name']).size().reset_index(name="count")

chess_black_openings = chess_winner_opening_df[chess_winner_opening_df['winner'] == 'black']

chess_white_openings = chess_winner_opening_df[chess_winner_opening_df['winner'] == 'white']

# chess_best_white_openings = chess_white_openings.sort_values("count", ascending=False)

# chess_best_black_openings = chess_black_openings.sort_values("count", ascending=False)
```

```python
# get %

# chess_best_white_openings['percent'] = chess_best_white_openings[]

for i in range(len(chess_white_openings)-1):
    try:
        black_count = chess_black_openings[chess_black_openings['opening_name'] == chess_white_openings.iloc[i]['opening_name']]['count'][0]
        white_count = chess_white_openings.iloc[i]['count']

        percent = white_count / (white_count+black_count)

        chess_white_openings.iloc[i]['win_percent'] = percent
    
        print(percent)
    except:
        print('=x')
#     break
```
