### Conclusão

Assim, os testes de hipóteses mostraram que a abertura utilizada está muito ligada ao nível dos jogadores, uma vez que para cada grupo pré estabelecido podemos observar que poucas vezes coincidia a abertura com maior taxa de vitórias.
